#### Comparison between cycles and instructions with different Optimizers

```python
from turtle import color

w = 0.4
x = [df['Optimization'][i] for i in range(0, len(df))][::-1]

cycles = df['cycles:u'][::-1]
instructions = df['instructions:u'][::-1]
# l1_dcache_loads = df['L1-dcache-loads'][::-1]
# l1_dcache_loads_misses = df['L1-dcache-load-misses'][::-1]
#l1_dcache_loads_stores = df['L1-dcache-stores']

bar1 = np.arange(len(x))
bar2 = [i+w for i in bar1]
# bar3 = [i+w for i in bar2]
#bar4 = [i+w for i in bar3]
#bar5 = [i+w for i in bar4]

plt.barh(bar1, cycles, w, label="cycles:u", color="#264653")
plt.barh(bar2, instructions, w, label="instructions:u", color="#2A9D8F")
# plt.barh(bar3, l1_dcache_loads, w, label="L1-dcache-loads")
# plt.bar(bar4, l1_dcache_loads_misses, w, label="L1-dcache-load-misses")
#plt.bar(bar5, l1_dcache_loads_stores, w, label="l1_dcache_loads_stores")

plt.xlabel("Optimization")
plt.ylabel("benchmark")
plt.title("Comparison between cycles and instructions with different Optimizers")

plt.yticks(bar1+w, x)
plt.legend(loc='lower right')
plt.show()
```

#### Comparison between cache flow with different Optimizers

```python
w = 0.3
x = [df['Optimization'][i] for i in range(0, len(df))][::-1]

#cycles = df['cycles:u'][::-1]
#instructions = df['instructions:u'][::-1]
l1_dcache_loads = df['L1-dcache-loads'][::-1]
l1_dcache_loads_misses = df['L1-dcache-load-misses'][::-1]
l1_dcache_loads_stores = df['L1-dcache-stores']

bar1 = np.arange(len(x))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]
#bar4 = [i+w for i in bar3]
#bar5 = [i+w for i in bar4]

# plt.barh(bar1, cycles, w, label="cycles:u")
# plt.barh(bar2, instructions, w, label="instructions:u")
plt.barh(bar1, l1_dcache_loads, w, label="L1-dcache-loads", color="#264653")
plt.barh(bar2, l1_dcache_loads_misses, w, label="L1-dcache-load-misses", color="#2A9D8F")
plt.barh(bar3, l1_dcache_loads_stores, w, label="l1_dcache_loads_stores", color="#E9C46A")

plt.xlabel("Optimization")
plt.ylabel("benchmark")
plt.title("Comparison between cache flow with different Optimizers")

plt.yticks(bar1+w, x)
plt.legend(loc='center right')
plt.show()
```
